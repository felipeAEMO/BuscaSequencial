```
import cv2
import random
from google.colab.patches import cv2_imshow
import numpy

imagem = numpy.zeros(shape=(100,100))

print(imagem)

print("Antiga")

img = cv2.imread("IMAGEM_NOTICIA_5.jpg", 1) 
print('Original Dimensions : ',img.shape)
width = 50
height = 50
dim = (width, height)
cv2_imshow(img)

#Algoritmo for zoom 
def zoom(img, value):
    value = random.uniform(value, 1)
    height, width = img.shape[:2]
    h_taken = int(value*height)
    w_taken = int(value*width)
    h_start = random.randint(0, height-h_taken)
    w_start = random.randint(0, width-w_taken)
    img = img[h_start:h_start+h_taken, w_start:w_start+w_taken, :]
    return img




print("Data augmentation:")



imgZoom = zoom(img, 0.5)
cv2_imshow(imgZoom)
```
